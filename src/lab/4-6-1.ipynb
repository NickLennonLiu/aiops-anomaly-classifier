```python
dt_path = "../data/system-a"
import os
import csv

cmdb_dict = {}
row_cnt = []
for file in os.listdir(dt_path):
    file_path = os.path.join(dt_path, file)
    cmdb_id, kpi_name = os.path.splitext(file)[0].split('##')
    if cmdb_id not in cmdb_dict:
        cmdb_dict[cmdb_id] = [kpi_name]
    else:
        cmdb_dict[cmdb_id].append(kpi_name)
    continue
    with open(file_path) as f:
        reader = csv.reader(f)
        for row in reader:
            pass
        row_cnt.append(reader.line_num)

print(sum([len(i) for i in cmdb_dict.values()]))
print(len(cmdb_dict.values()))
```
